## ***Is the dataset balanced?***
*   **No, the dataset is imbalanced.**
*   **"Good"** customers constitute *70.0%* of the data, whereas **"Bad"** customers account for only *30.0%*.
*   This imbalance is significant as it may lead our future prediction model to be biased towards the majority class (**"Good"**). Consequently, it will excel at predicting "Good" customers but will struggle to accurately identify the risky ***"Bad"** customers.
